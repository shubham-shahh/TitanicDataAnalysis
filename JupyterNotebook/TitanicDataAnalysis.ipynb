```python
df.dropna(inplace=True)
```

```python
sex = pd.get_dummies(df['Sex'], drop_first=True)
```

```python
embark = pd.get_dummies(df['Embarked'], drop_first=True)
```

```python
Pclass = pd.get_dummies(df['Pclass'], drop_first=True)
```

```python
df = pd.concat([df, sex, embark, Pclass], axis = 1)
```

```python
df
```

```python
kk = df.drop(columns = ['Cabin'])
```

```python
kk = df.drop(columns = ['Name'])
kk = kk.drop(columns = ['PassengerId'])
kk = kk.drop(columns = ['Sex'])
kk = kk.drop(columns = ['Embarked'])
kk = kk.drop(columns = ['Ticket'])
kk = kk.drop(columns = ['Cabin'])
kk = kk.drop(columns = ['Survived'])
kk

y = df['Survived']
y
```

## Train

```python
x_train, x_test, y_train, y_test = train_test_split(kk, y, test_size=0.2, random_state=1)
```

```python
model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
classification_report(y_test,predictions)
```
